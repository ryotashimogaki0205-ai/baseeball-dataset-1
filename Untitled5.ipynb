````python
import sys
# !{sys.executable} -m pip install streamlit # Streamlit is already installed, no need to run again.
import streamlit as st
import pandas as pd
import io # For handling in-memory files for download

# --- Streamlit Session State Initialization ---
# Initialize feedback_data in session state if it doesn't exist
if 'feedback_data' not in st.session_state:
    st.session_state.feedback_data = []

# --- Feedback Section ---
st.title('フィードバック')

# ユーザーがフィードバックコメントを入力するためのテキストエリア
feedback_comment = st.text_area('ご意見・ご感想をお願いします。', key='feedback_text')

# ユーザーが評価を選択するためのラジオボタンのグループ
feedback_rating = st.radio(
    'このアプリの満足度はいかがでしたか？',
    ['満足', '普通', '不満'],
    key='feedback_rating'
)

# フィードバックを送信するためのボタン
submit_button = st.button('フィードバックを送信', key='submit_feedback')

# ボタンが押されたときの処理
if submit_button:
    if feedback_comment or feedback_rating:
        new_feedback_entry = {
            'comment': feedback_comment,
            'rating': feedback_rating,
            'timestamp': pd.Timestamp.now()
        }
        st.session_state.feedback_data.append(new_feedback_entry)
        st.success('フィードバックを送信しました！ご協力ありがとうございます。')
    else:
        st.warning('フィードバックコメントまたは満足度を入力してください。')

# --- Collected Feedback Display ---
st.subheader('これまでに集まったフィードバック')
if st.session_state.feedback_data:
    feedback_df = pd.DataFrame(st.session_state.feedback_data)
    st.dataframe(feedback_df)

    # ダウンロードボタン
    csv_buffer = io.StringIO()
    feedback_df.to_csv(csv_buffer, index=False)
    st.download_button(
        label="フィードバックをCSVでダウンロード",
        data=csv_buffer.getvalue(),
        file_name="feedback_data.csv",
        mime="text/csv"
    )
else:
    st.info('まだフィードバックはありません。')
````

````python
app_code = """import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib # 日本語表示のためにインポート
import matplotlib.patches as patches
import io # For handling in-memory files for download

# --- Streamlit Session State Initialization ---
# Initialize df in session state if it doesn't exist
if 'df' not in st.session_state:
    st.session_state.df = pd.DataFrame(columns=['pitch_x', 'pitch_y', 'ball_direction', 'exit_velocity', 'pitch_type'])
    # Add initial data only once if the DataFrame is newly created and empty
    initial_entries = [
        {
            'pitch_x': -0.8,
            'pitch_y': 2.0,
            'ball_direction': 'Left',
            'exit_velocity': 108,
            'pitch_type': 'Fastball'
        },
        {
            'pitch_x': 0.1,
            'pitch_y': 3.1,
            'ball_direction': 'Center',
            'exit_velocity': 98,
            'pitch_type': 'Curveball'
        }
    ]
    if st.session_state.df.empty and initial_entries:
        st.session_state.df = pd.concat([st.session_state.df, pd.DataFrame(initial_entries)], ignore_index=True)

# Initialize feedback_data in session state if it doesn't exist
if 'feedback_data' not in st.session_state:
    st.session_state.feedback_data = []

# --- Streamlit Application Title ---
st.title('打席データ分析とフィードバックアプリ')

# --- 2. 新しい打席データを入力 (Streamlit UI) ---
st.header('新しい打席データを入力')

with st.form("new_batting_data_form"):
    col1, col2 = st.columns(2)
    with col1:
        input_pitch_x = st.number_input('投球の左右位置 (pitch_x)', min_value=-3.0, max_value=3.0, value=0.0, step=0.1, key='input_pitch_x')
        input_pitch_y = st.number_input('投球の高さ位置 (pitch_y)', min_value=-5.0, max_value=5.0, value=2.5, step=0.1, key='input_pitch_y')
        input_exit_velocity = st.number_input('打球速度 (mph)', min_value=50, max_value=120, value=100, step=1, key='input_exit_velocity')
    with col2:
        input_ball_direction = st.selectbox('打球方向', ['Left', 'Center', 'Right', 'Grounder', 'Flyball'], key='input_ball_direction')
        input_pitch_type = st.selectbox('球種', ['Fastball', 'Curveball', 'Slider', 'Changeup', 'Other'], key='input_pitch_type')

    add_data_button = st.form_submit_button('この打席データを追加')

    if add_data_button:
        new_entry = {
            'pitch_x': input_pitch_x,
            'pitch_y': input_pitch_y,
            'ball_direction': input_ball_direction,
            'exit_velocity': input_exit_velocity,
            'pitch_type': input_pitch_type
        }
        new_df_to_add = pd.DataFrame([new_entry])
        st.session_state.df = pd.concat([st.session_state.df, new_df_to_add], ignore_index=True)
        st.success('新しい打席データが追加されました！')

# --- 3. データフレームの全体確認 ---
st.header('現在のデータフレーム')
if not st.session_state.df.empty:
    st.dataframe(st.session_state.df)
else:
    st.info('まだデータがありません。新しいデータを入力してください。')

# --- 4. 各分析グラフの表示 ---
st.header('分析グラフ')

if not st.session_state.df.empty:
    # 4.1 投球位置と球種
    st.subheader('投球位置と球種')
    fig1, ax1 = plt.subplots(figsize=(8, 10)) # 縦長の長方形に変更
    sns.scatterplot(x='pitch_x', y='pitch_y', hue='pitch_type', data=st.session_state.df, s=100, alpha=0.7, ax=ax1)
    ax1.set_title('投球位置と球種')
    ax1.set_xlabel('投球の左右位置 (捕手視点など)')
    ax1.set_ylabel('投球の高さ位置 (地面からなど)')
    ax1.grid(True)
    ax1.set_xlim(-2.0, 2.0)
    ax1.set_ylim(-4.0, 4.0)
    strike_zone1 = patches.Rectangle((-1.5, -2.5), 3.0, 5.0, linewidth=3, edgecolor='red', facecolor='none', linestyle='--')
    ax1.add_patch(strike_zone1)
    ax1.set_aspect('equal', adjustable='box')
    st.pyplot(fig1)
    plt.close(fig1) # Close the figure to free up memory

    # 4.2 打球方向ごとの打球速度
    st.subheader('打球方向ごとの打球速度')
    fig2, ax2 = plt.subplots(figsize=(8, 6))
    sns.boxplot(x='ball_direction', y='exit_velocity', data=st.session_state.df, ax=ax2)
    ax2.set_title('打球方向ごとの打球速度')
    ax2.set_xlabel('打球方向')
    ax2.set_ylabel('打球速度 (mph)')
    ax2.grid(axis='y')
    st.pyplot(fig2)
    plt.close(fig2) # Close the figure to free up memory

    # 4.3 打球方向ごとの投球位置と球種 (ファセットグリッド)
    st.subheader('打球方向ごとの投球位置と球種 (ファセットグリッド)')
    g = sns.relplot(x='pitch_x', y='pitch_y', hue='pitch_type', col='ball_direction', data=st.session_state.df, kind='scatter', s=100, alpha=0.7, col_wrap=3, height=5, aspect=0.8)
    g.set_axis_labels('投球の左右位置', '投球の高さ位置')
    g.set_titles(col_template='打球方向: {col_name}')
    g.fig.suptitle('打球方向ごとの投球位置と球種', y=1.02) # 全体タイトル
    for ax in g.axes.flat:
        strike_zone_facet = patches.Rectangle((-1.5, -2.5), 3.0, 5.0, linewidth=1, edgecolor='red', facecolor='none', linestyle='--')
        ax.add_patch(strike_zone_facet)
        ax.set_xlim(-2.0, 2.0)
        ax.set_ylim(-4.0, 4.0)
        ax.set_aspect('equal', adjustable='box')
    plt.tight_layout()
    st.pyplot(g.fig)
    plt.close(g.fig) # Close the figure to free up memory
else:
    st.info('データがないためグラフを表示できません。データを追加してください。')

# --- Feedback Section (Integrated from previous subtask) ---
st.header('フィードバック')

# ユーザーがフィードバックコメントを入力するためのテキストエリア
feedback_comment = st.text_area('ご意見・ご感想をお願いします。', key='feedback_text')

# ユーザーが評価を選択するためのラジオボタンのグループ
feedback_rating = st.radio(
    'このアプリの満足度はいかがでしたか？',
    ['満足', '普通', '不満'],
    key='feedback_rating'
)

# フィードバックを送信するためのボタン
submit_feedback_button = st.button('フィードバックを送信', key='submit_feedback')

# ボタンが押されたときの処理
if submit_feedback_button:
    if feedback_comment or feedback_rating:
        new_feedback_entry = {
            'comment': feedback_comment,
            'rating': feedback_rating,
            'timestamp': pd.Timestamp.now()
        }
        st.session_state.feedback_data.append(new_feedback_entry)
        st.success('フィードバックを送信しました！ご協力ありがとうございます。')
        # Clear input fields after submission (optional, but good UX)
        # Note: Clearing text_area usually requires a callback or re-running the app with an empty value
        # For simplicity, we just show success message for now.
    else:
        st.warning('フィードバックコメントまたは満足度を入力してください。')

# --- Collected Feedback Display ---
st.subheader('これまでに集まったフィードバック')
if st.session_state.feedback_data:
    feedback_df = pd.DataFrame(st.session_state.feedback_data)
    st.dataframe(feedback_df)

    # ダウンロードボタン
    csv_buffer = io.StringIO()
    feedback_df.to_csv(csv_buffer, index=False)
    st.download_button(
        label="フィードバックをCSVでダウンロード",
        data=csv_buffer.getvalue(),
        file_name="feedback_data.csv",
        mime="text/csv"
    )
else:
    st.info('まだフィードバックはありません。')
"""

with open('main_app.py', 'w') as f:
    f.write(app_code)

print("main_app.py has been created.")
````

### Subtask:

#### Instructions
1. Streamlit Cloudのダッシュボードにログインします。
2. 「Deploy an app」または「New app」ボタンをクリックします。

7. 「Python version」は通常デフォルトで問題ありませんが、必要に応じて選択します。

## GitHub ActionsによるCI/CD導入の説明

GitHub Actionsを利用したCI/CDパイプラインの導入方法について説明します。CI/CD (継続的インテグレーション/継続的デリバリー) を導入することで、コードの変更が自動的にテストされ、問題がなければStreamlit Cloudなどのデプロイ環境へ自動的にデプロイされるようになります。

### GitHub Actionsの概念
*   **ワークフロー (Workflow)**: 1つ以上のジョブを含む自動化されたプロセスです。`.github/workflows` ディレクトリにYAMLファイルとして定義します。
*   **イベント (Event/Trigger)**: ワークフローを実行するきっかけとなるGitHubのアクティビティです。例: `push` (コードがリポジトリにプッシュされた時)、`pull_request` (プルリクエストが作成された時) など。
*   **ジョブ (Job)**: ワークフロー内で並列または連続して実行される一連のステップです。通常、特定の環境（例: Ubuntuの仮想マシン）上で実行されます。
*   **ステップ (Step)**: ジョブ内の個々のアクションです。コマンドの実行、スクリプトの実行、またはGitHub Marketplaceから利用可能なアクションの使用などが含まれます。
*   **アクション (Action)**: 特定のタスクを実行するための再利用可能なコンポーネントです。例: `actions/checkout@v4` (リポジトリをチェックアウトする)、`actions/setup-python@v5` (Python環境をセットアップする) など。

### CI/CDパイプラインの例 (Streamlit Cloudへのデプロイ)

以下は、`main`ブランチへのプッシュ時にPythonの依存関係をインストールし、テストを実行し、問題がなければStreamlit CloudへデプロイするGitHub Actionsワークフローの例です。

`.github/workflows/main.yml`

```yaml
name: Streamlit CI/CD

on:
  push:
    branches:
      - main
  pull_request:
    branches:
      - main

jobs:
  build-and-deploy:
    runs-on: ubuntu-latest

    steps:
    - name: 🔄 Checkout repository
      uses: actions/checkout@v4

    - name: 🐍 Set up Python
      uses: actions/setup-python@v5
      with:
        python-version: '3.9' # またはお使いのPythonバージョンに合わせてください

    - name: 📦 Install dependencies
      run: |
        python -m pip install --upgrade pip
        pip install -r requirements.txt

    # - name: 🧪 Run tests (オプション)
    #   run: |
    #     # ここにテストコマンドを記述します
    #     # 例: pytest
    #     echo "No tests configured yet."

    - name: 🚀 Deploy to Streamlit Cloud
      if: github.ref == 'refs/heads/main' # mainブランチへのプッシュ時のみデプロイを実行
      uses: streamlit/streamlit-deploy@v0.1.0
      with:
        streamlit_app_path: main_app.py
        streamlit_api_key: ${{ secrets.STREAMLIT_API_KEY }}
        # 他のStreamlit Cloudのデプロイ設定（例: GITHUB_TOKEN, REPOSITORY）は、
        # Streamlit Cloud側の設定でGitHubリポジトリと連携していれば不要な場合があります。
        # 詳細はStreamlit Cloudのドキュメントを参照してください。
```

## GitHub ActionsによるCI/CD導入の説明

GitHub Actionsを利用したCI/CDパイプラインの導入方法について説明します。CI/CD (継続的インテグレーション/継続的デリバリー) を導入することで、コードの変更が自動的にテストされ、問題がなければStreamlit Cloudなどのデプロイ環境へ自動的にデプロイされるようになります。

### GitHub Actionsの概念
*   **ワークフロー (Workflow)**: 1つ以上のジョブを含む自動化されたプロセスです。`.github/workflows` ディレクトリにYAMLファイルとして定義します。
*   **イベント (Event/Trigger)**: ワークフローを実行するきっかけとなるGitHubのアクティビティです。例: `push` (コードがリポジトリにプッシュされた時)、`pull_request` (プルリクエストが作成された時) など。
*   **ジョブ (Job)**: ワークフロー内で並列または連続して実行される一連のステップです。通常、特定の環境（例: Ubuntuの仮想マシン）上で実行されます。
*   **ステップ (Step)**: ジョブ内の個々のアクションです。コマンドの実行、スクリプトの実行、またはGitHub Marketplaceから利用可能なアクションの使用などが含まれます。
*   **アクション (Action)**: 特定のタスクを実行するための再利用可能なコンポーネントです。例: `actions/checkout@v4` (リポジトリをチェックアウトする)、`actions/setup-python@v5` (Python環境をセットアップする) など。

name: Streamlit CI/CD

on:
  push:
    branches:
      - main
  pull_request:
    branches:
      - main

jobs:
  build-and-deploy:
    runs-on: ubuntu-latest

    steps:
    - name: 🔄 Checkout repository
      uses: actions/checkout@v4

    - name: 🐍 Set up Python
      uses: actions/setup-python@v5
      with:

    #   run: |
    #     # ここにテストコマンドを記述します
    #     # 例: pytest
    #     echo "No tests configured yet."

    - name: 🚀 Deploy to Streamlit Cloud
      if: github.ref == 'refs/heads/main' # mainブランチへのプッシュ時のみデプロイを実行
      uses: streamlit/streamlit-deploy@v0.1.0
      with:
        streamlit_app_path: main_app.py
        streamlit_api_key: ${{ secrets.STREAMLIT_API_KEY }}
        # 他のStreamlit Cloudのデプロイ設定（例: GITHUB_TOKEN, REPOSITORY）は、
